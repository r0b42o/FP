Berechne Atomgröße Graphit

```python
Richtung1 = np.array([1.209, 1.214, 1.192, 1.195, 1.212, 1.188, 1.218, 1.203, 1.202, 1.215])
Richtung2 = np.array([1.301, 1.286, 1.277, 1.280, 1.283])
Richtung3 = np.array([1.348, 1.333, 1.350, 1.342, 1.361, 1.343, 1.343, 1.366, 1.357, 1.344])

FehlerR1 = 0.098/4
FehlerR2 = 0.103/4
FehlerR3 = 0.087/4

FehlerR1_array = np.full(len(Richtung1), FehlerR1)
FehlerR2_array = np.full(len(Richtung2), FehlerR2)
FehlerR3_array = np.full(len(Richtung3), FehlerR3)

Richtung1 = unp.uarray(Richtung1, FehlerR1_array)
Richtung2 = unp.uarray(Richtung2, FehlerR2_array)
Richtung3 = unp.uarray(Richtung3, FehlerR3_array)

nominal_values = unp.nominal_values(Richtung1)/10
stdR1 = np.std(nominal_values)
Fehler2R1 = ufloat(0, stdR1)

nominal_values = unp.nominal_values(Richtung2)/10
stdR2 = np.std(nominal_values)
Fehler2R2 = ufloat(0, stdR2)

nominal_values = unp.nominal_values(Richtung3)/10
stdR3 = np.std(nominal_values)
Fehler2R3 = ufloat(0, stdR3)

LängeAtomR1 = np.mean(Richtung1)/10 + Fehler2R1
LängeAtomR2 = np.mean(Richtung2)/10 + Fehler2R2
LängeAtomR3 = np.mean(Richtung3)/10 + Fehler2R3

print(LängeAtomR1)
print(LängeAtomR2)
print(LängeAtomR3)

gittverktorlange = (LängeAtomR1 + LängeAtomR2 + LängeAtomR3)/3
print(gittverktorlange)
```

Berechne nanopartikelgröße

```python
N1 = np.array([10.9, 11.8])
N2 = np.array([10.4, 12.6])
N3 = np.array([8.8, 12.4])
N4 = np.array([9.2, 8.2])
N5 = np.array([8.1, 10.9])
N6 = np.array([9.2, 10.8])
N7 = np.array([8.0, 10.4])
N8 = np.array([8.4, 11.3])
N9 = np.array([8.4, 10.5])
N10 = np.array([8.0, 10.4])

R = np.array([np.mean(N1), np.mean(N2), np.mean(N3), np.mean(N4), np.mean(N5), np.mean(N6), np.mean(N7), np.mean(N8), np.mean(N9), np.mean(N10)])
RFehler = np.std(R)
R = np.mean(R)
R = ufloat(R, RFehler)/2
print(R)
```
